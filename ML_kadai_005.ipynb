```python
#1matplotlib.pyplotメソッドとnumpyをインポートするコードを記述する。
import matplotlib.pyplot as plt
import numpy as np

#2 6個のグラフを表示する領域を作成します。
#Figure(描画領域全体)とAxes(ひとつひとつのプロットを描く領域)を作成するコードを記述する。
fig, axes = plt.subplots(2,3, tight_layout = True)

y0 = [1,2,-5,2]
axes[0,0].plot(y0)

x1 = np.linspace(0,10,100)
y1 = 2 + 2 * np.sin(2 * x1)

axes[0,1].plot(x1, y1, linewidth = 2.0 )

x2 = np.random .normal(15, 5, 2000)
axes[0,2].hist(x2, bins=30)

np.random.seed(3)
x3 = 4 + np.random.normal(0, 2, 60)
y3 = 4 + np.random.normal(0, 2, len(x3))

# プロットする
axes[1,0].scatter(x3, y3)

plt.show()

```
